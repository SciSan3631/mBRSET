```python
import os
import pandas as pd
```

```python
df = pd.read_csv('labels_mbrset.csv')
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt


# Count the occurrences of each artifact
artifact_counts = df['final_artifacts'].value_counts()

# Create the pie chart
plt.pie(artifact_counts.values, labels=artifact_counts.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.title('Distribution of Final Artifacts')
plt.show()

# Count the occurrences of each artifact
artifact_counts = df['final_artifacts'].value_counts()

# Create the bar chart
plt.bar(artifact_counts.index, artifact_counts.values)
plt.xlabel('Final Artifacts')
plt.ylabel('Count')
plt.title('Distribution of Final Artifacts')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'final_quality' and 'final_artifacts' are your column names
cross_table = pd.crosstab(df['final_quality'], df['final_artifacts'])

# Create a heatmap
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.heatmap(cross_table, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Cross-Tabulation of Final Quality and Final Artifacts")
plt.xlabel("Final Artifacts")
plt.ylabel("Final Quality")
plt.show()
```
